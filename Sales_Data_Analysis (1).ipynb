```python
# PRODUCT WISE SALES CATEGORY

top_product = df.groupby( 'Product_Category')['Sales_Amount'].sum()
top_product.sort_values(ascending = False)
```

```python
# REGION WISE SALES

region_wise_sales = df.groupby( 'Region')['Sales_Amount'].sum()
region_wise_sales.sort_values(ascending = False)
```

```python
# REGION WISE SALES CATEGORY

region_wise_sales_category = df.groupby('Product_Category')['Sales_Amount'].sum()
region_wise_sales_category.sort_values(ascending = False)
```

```python
# BEST SALES REPRESENTATTIVE

sales_rep_sales = df.groupby('Sales_Rep')['Sales_Amount'].sum()
sales_rep_sales.sort_values(ascending = False)
```

```python
import matplotlib.pyplot as plt
top_product.plot(kind = "bar")
plt.title("Top Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
region_wise_sales.plot(kind = "bar",color = "green")
plt.title("Region wise sales")
plt.xlabel("region")
plt.ylabel("Sales")
plt.show()
```

```python
region_wise_sales_category.plot(kind = "bar",color = "orange")
plt.title("Region wise sales category")
plt.xlabel("region")
plt.ylabel("Sales category")
plt.show()
```

```python

sales_rep_sales .plot(kind = "bar")
plt.title("Best Sales Representative")
plt.xlabel("Sales_rep")
plt.ylabel("Sales category")
plt.show()
```
