```python
# Data Operations
data = [
    [25, "IT", 50000, "Mumbai", "Yes"],
    [30, np.nan,60000, "Delhi", "No"],
    [np.nan, "HR", 55000, "Pune", "Yes"],
    [28, "Finance", np.nan, "Bangalore", "No"],
    [35, "IT", 70000, np.nan, "Yes"],
    [40, "Sales", 80000, "Chennai", np.nan]
]

df = pd.DataFrame(
    data,
    index=["Row 1", "Row 2", "Row 3", "Row 4", "Row 5", "Row 6"],
    columns=["Age", "Department", "Salary", "City", "Active"]
)

df    # with table
```

```python
data = [
    [25, "IT", 50000, "Mumbai", "Yes"],
    [30, np.nan,60000, "Delhi", "No"],
    [np.nan, "HR", 55000, "Pune", "Yes"],
    [28, "Finance", np.nan, "Bangalore", "No"],
    [35, "IT", 70000, np.nan, "Yes"],
    [40, "Sales", 80000, "Chennai", np.nan]
]

df = pd.DataFrame(
    data,
    index=["Row 1", "Row 2", "Row 3", "Row 4", "Row 5", "Row 6"],
    columns=["Age", "Department", "Salary", "City", "Active"]
)

print(df)      # without table
```
